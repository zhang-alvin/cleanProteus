```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
directory="./"
!ls
```

```python
time = np.genfromtxt(directory+"timeHistory.txt", dtype=float)
p_force = np.genfromtxt(directory+"forceHistory_p.txt", dtype=float)
v_force = np.genfromtxt(directory+"forceHistory_v.txt", dtype=float)
```

```python
force=p_force+v_force
```

```python
print(force.shape)
```

```python
N = force.shape[0]
```

```python
plt.plot(time, force[:,0], 'b', time,force[:,1], 'y')
#plt.axis([0, N, -1, 1])
#plt.show()
```

```python
D = 0.1
U = 1.0
rho=1.0
Cd = force[:,0]*2/rho/D/U/U
Cl = force[:,1]*2/rho/D/U/U
```

```python
plt.figure()
p1, = plt.plot(range(N), Cd, 'b', label='Cd')
p2, = plt.plot(range(N), Cl, 'y', label='Cl')
plt.axis([0, N, -1, 4])
plt.legend(handles=[p1,p2])
#plt.show()
```
