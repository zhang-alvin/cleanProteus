```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
!pwd
```

```python
directory="./"
```

```python
force =np.genfromtxt(directory+"particle_forceHistory.txt", dtype=float)
```

```python
N = force.shape[0]
```

```python
plt.figure()
plt.plot(range(N), force[:,0], 'b', range(N),force[:,1], 'y')
plt.axis([0, N, -1, 1])
plt.show()
```

```python
D = 0.1
U = 1.0
rho=1.0
Cd = force[:,0]*2/rho/D/U/U
Cl = force[:,1]*2/rho/D/U/U
```

```python
plt.figure()
p1, = plt.plot(range(N), Cd, 'b', label='Cd')
p2, = plt.plot(range(N), Cl, 'y', label='Cl')
plt.axis([0, N, -1, 4])
plt.legend(handles=[p1,p2])
plt.show()
```
